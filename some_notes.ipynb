```python
# Filepath
gas_outputs_file_path = '.\derived_gas_outputs_minutes.csv'
# Store data in DataFrame and provide some lines
gas_data = pd.read_csv(gas_outputs_file_path, index_col='timestamp', parse_dates=['timestamp'])
gas_data.drop(gas_data.columns[0], inplace=True, axis=1)
gas_data.head()
```

```python
gas_data['sum'] = gas_data['mean_gas_used'] + gas_data['mean_gas_refund'] + gas_data['mean_gas_burned']
gas_data['compare'] = gas_data['sum']-gas_data['mean_gas_limit']
gas_data.describe()
```

```python
# Some values can be obtained from existing columns. See documentation (ADD CITATION)
# GasPremium = GasFeeCap-BaseFee (see https://filecoin.io/blog/posts/filecoin-features-gas-fees/), then, equivalently,
gas_data['mean_base_fee']=gas_data['mean_gas_fee_cap']-gas_data['mean_gas_premium']
# BaseFeeBurned = BaseFee * GasUsage, then, equivalently,
gas_data['mean_base_fee_burn_2'] = gas_data['mean_base_fee'] * gas_data['mean_gas_used']
gas_data['mean_base_fee_burn_3'] = gas_data['mean_parent_base_fee'] * gas_data['mean_gas_used']
# We can check if mean_base_fee_burn and mean_base_fee_burn_2 coincide:
gas_data['coincidence_1'] = gas_data['mean_base_fee_burn_2'] / gas_data['mean_base_fee_burn']
gas_data['coincidence_2'] = gas_data['mean_base_fee_burn_3'] / gas_data['mean_base_fee_burn']

# GasNotUsed = GasLimit - GasUsed, then, equivalently,
gas_data['mean_gas_not_used'] = gas_data['mean_gas_limit'] - gas_data['mean_gas_used']
# OverestimationBurned = SomeProportion * GasNotUsed, then, equivalently,
# gas_data['mean_some_proportion'] = gas_data['mean_over_estimation_burn'] / gas_data['mean_gas_not_used']
# Refund_in_attoFIL = Premium * GasNotUsed - OverEstimationBurn, then, equivalently,
gas_data['mean_gas_refund_in_attoFIL'] = gas_data['mean_gas_premium'] * gas_data['mean_gas_not_used'] - gas_data['mean_over_estimation_burn']
# We can check if mean_gas_refund and mean_gas_refund_2 coincide: NO; DIFFERENT UNITS
#gas_data['coincidence_2'] = gas_data['mean_gas_refund'] - gas_data['mean_gas_refund_2']

# GasBurned = BaseFeeBurned + OverEstimationBurned, then, equivalently,
gas_data['mean_gas_burned_in_attoFIL'] = gas_data['mean_base_fee_burn'] + gas_data['mean_over_estimation_burn']
# We can check if mean_gas_burned and mean_gas_burned_2 coincide: NO; DIFFERENT UNITS
#gas_data['coincidence_3'] = gas_data['mean_gas_burned'] - gas_data['mean_gas_burned_2']
```

## Data visualization

```python
gas_data['mean_gas_fee_cap'].asfreq('d').plot() # asfreq method is used to convert a time series to a specified frequency. Here it is daily frequency.
plt.title('Mean base fee cap over time(daily frequency)')
plt.show()
```

```python
gas_data['mean_gas_fee_cap'].asfreq('d').plot() # asfreq method is used to convert a time series to a specified frequency. Here it is daily frequency.
plt.title('Mean base fee cap over time(daily frequency)')
plt.show()
```

```python
gas_data.asfreq('D').plot(subplots=True, figsize=(10,30))
plt.title('Gas outputs attributes')
plt.show()
```

```python
gas_daily = gas_data.resample('d').mean()
```

```python
gas_daily.plot(subplots=True, figsize=(10,30))
plt.title('Gas outputs attributes')
plt.show()
```

Gas Daily

```python
# Resample taking mean to a daily basis to make it more clear
gas_daily = gas_data.resample('d').mean()

gasunit_features = ['mean_gas_limit','mean_gas_used','mean_gas_refund','mean_gas_burned']
gas_daily_df = gas_daily[gasunit_features]
gas_daily_df.head()
```

```python
# We analyse now the evolution of fees per GasUnit, in particular: GasFeeCap, GasPremum and ParentBaseFee
feepergasunit_features = ['mean_gas_fee_cap','mean_gas_premium','mean_parent_base_fee']

fpgu_daily_df = gas_daily[feepergasunit_features]
fpgu_daily_df.plot(title='Fees per GasUnit', ylabel='Fees', figsize=(16,5))
plt.show()
```

Gas Daily using two features
